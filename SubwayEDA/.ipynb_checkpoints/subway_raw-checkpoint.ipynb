```python
#2019.01_06중에 언제 지하철을 가장 많이 이용했을까?(기준:승하차총승객수)
import pandas as pd
```

```python
subway_month_201901 = pd.read_csv('CARD_SUBWAY_MONTH_201901.csv')
subway_month_201902 = pd.read_csv('CARD_SUBWAY_MONTH_201902.csv')
subway_month_201903 = pd.read_csv('CARD_SUBWAY_MONTH_201903.csv')
subway_month_201904 = pd.read_csv('CARD_SUBWAY_MONTH_201904.csv')
subway_month_201905 = pd.read_csv('CARD_SUBWAY_MONTH_201905.csv')
subway_month_201906 = pd.read_csv('CARD_SUBWAY_MONTH_201906.csv')

raw = pd.concat([subway_month_201901,subway_month_201902,subway_month_201903\
                ,subway_month_201904,subway_month_201905,subway_month_201906\
               ],axis=0)

```

```python
raw
```

```python
raw['사용일자'] = pd.to_datetime(raw['사용일자'], format='%Y%m%d')
```

```python
raw['요일'] = raw['사용일자'].dt.dayofweek
```

```python
conv = {
    0 : '일',
    1 : '월',
    2 : '화',
    3 : '수',
    4 : '목',
    5 : '금',
    6 : '토'
}
```

```python
raw['요일'] = raw['요일'].map(conv)
```

```python
raw.sample(5)
```

```python
raw['연월'] = raw['사용일자'].dt.strftime('%Y-%m')
```

```python
raw['월일'] = raw['사용일자'].dt.strftime('%m-%d')
```

```python
raw['승하자총승객수'] = raw['승차총승객수'] + raw['하차총승객수']
```

```python
raw = raw[['사용일자', '요일', '노선명', '역명', '승차총승객수', '하차총승객수', '승하자총승객수', '연월', '월일', '등록일자' ]]
```
