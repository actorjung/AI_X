```python
#연월 각각에 대해 일자별(월일별)승하차총승객수 그래프 그려 볼까요?(pointplot)
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
subway_raw = pd.read_csv('subway_raw.csv')
```

```python
#한글폰트깨짐 해결
import matplotlib.font_manager as fm

font_list = [font.name for font in fm.fontManager.ttflist]
font_list
plt.rcParams['font.family'] = 'HYGungSo-Bold'
```

```python
df_subway = subway_raw.groupby(['연월','월일'])\
                    .agg(승하차총승객수 = ('승하자총승객수','sum'))
df_subway
```

```python
ax = sns.lineplot(data = df_subway, x = '월일', y = '승하차총승객수')

plt.xticks(rotation=45)
```
