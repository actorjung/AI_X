```python
#가장 승객이 많이 타는 승차역은?
import pandas as pd
import numpy as np
import seaborn as sns
subway_raw = pd.read_csv('subway_raw.csv')
```

```python
subway_raw 
```

```python
df_subway = subway_raw.groupby('역명')\
                    .agg(승차총승객수 = ('승차총승객수','sum'))\
                    .sort_values('승차총승객수',ascending = False)
df_subway
```

```python
sns.lineplot(data = df_subway, x = '역명', y = '승차총승객수')
```
